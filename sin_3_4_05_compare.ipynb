```python
csv_fn = '/Users/chenghunglin/Documents/sound_absorption_test/sin_3_4_05_FEM.csv'
sin_3_4_FEM = pd.read_csv(csv_fn, sep=',', names=['frq', 'abs'], skiprows=1)
```

```python
fn = glob.glob('/Users/chenghunglin/Documents/sound_absorption_test/sin_3_4_05_rad_v1_0**.xlsx')
fn.sort()
print(*fn, sep = '\n')
plt.figure()
for f in fn:
    d = pd.read_excel(f, sheet_name='Constants')
    df = pd.read_excel(f, sheet_name='Sound_absorption', names=['Hz', 'abs'])
    f_step = df['Hz'][1]-df['Hz'][0]
    label = f'frequence step = {f_step:03d} Hz; Segments = {d['Value'][8]:03d}'
    plt.plot(df['Hz'], df['abs'], label=label)
plt.plot(sin_3_4_FEM['frq']*1000, sin_3_4_FEM['abs'], '.', label=f'sin, p=3, q=4, $\\theta$=0.5, FEM')
plt.legend()
plt.title(f'Sahpe = sin, p=3, q=4, $\\theta$=0.5')
```
