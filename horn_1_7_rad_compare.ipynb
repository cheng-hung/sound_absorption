```python
csv_fn = '/Users/chenghunglin/Documents/sound_absorption_test/horn_1_7_FEM.csv'
horn_1_7_FEM = pd.read_csv(csv_fn, sep=',', names=['frq', 'abs'], skiprows=1)
```

```python
fn = glob.glob('/Users/chenghunglin/Documents/sound_absorption_test/horn_1_7_rad_v1_1**.xlsx')
fn.sort()
print(*fn, sep = '\n')
plt.figure()
for f in fn:
    d = pd.read_excel(f, sheet_name='Constants')
    df = pd.read_excel(f, sheet_name='Sound_absorption', names=['Hz', 'abs'])
    f_step = df['Hz'][1]-df['Hz'][0]
    label = f'frequence step = {f_step:03d} Hz; Segments = {d['Value'][8]:03d}'
    plt.plot(df['Hz'], df['abs'], label=label)
plt.plot(horn_1_7_FEM['frq']*1000, horn_1_7_FEM['abs'], '.', label='horn, p=1, q=7, FEM')
plt.legend()
plt.title('Shape = horn, p = 1, q = 7')
```

```python
fn_rad = glob.glob('/Users/chenghunglin/Documents/sound_absorption_test/horn_1_7_rad_v1_1**.xlsx')
fn_rad.sort()
fn_vol = glob.glob('/Users/chenghunglin/Documents/sound_absorption_test/horn_1_7_vol_v1_0**.xlsx')
fn_vol.sort()
```
